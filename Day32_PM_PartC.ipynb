## Q1 — 1000 Trees vs 100 Trees

If a Random Forest with 1000 trees has the same accuracy as one with 100 trees, using 1000 trees is usually unnecessary.

**Tradeoffs:**
- **Compute Cost:** More trees increase training time
- **Prediction Latency:** Slower predictions in production
- **Marginal Improvement:** Little to no gain after a certain number of trees
- **Production Impact:** Higher infrastructure cost

**Conclusion:** Use 100 trees if performance is similar, as it is more efficient.

## Q2 — Coding: Compare Models with Cross Validation

````python
import pandas as pd
import time
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer, accuracy_score, f1_score

def compare_models(X, y, models_dict):
    results = []

    for name, model in models_dict.items():
        start = time.time()

        acc = cross_val_score(model, X, y, cv=5, scoring='accuracy')
        f1 = cross_val_score(model, X, y, cv=5, scoring='f1')

        end = time.time()

        results.append({
            "Model": name,
            "Accuracy_mean": acc.mean(),
            "Accuracy_std": acc.std(),
            "F1_mean": f1.mean(),
            "F1_std": f1.std(),
            "Training_time_sec": end - start
        })

    return pd.DataFrame(results)

print("Function ready. Pass models dictionary to use.")
````
